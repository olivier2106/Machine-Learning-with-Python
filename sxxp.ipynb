```python
columns_used = ['date', 
            'c2o', 
            'ratio_ma_5d_ma_20d', 'ratio_1d_ma_5d', 
            'ratio_volat_5d_volat_20d', 
            'rsi',
            'target', 'bips']
```

```python
data = []

for i in range(0, 5*600, 5):
    columns = ['date', 'last', 'open', 'close_1d', 'high', 'low']
    ta_one_stock = df_cleaned.iloc[:, [0,i+1, i +2, i+3, i+4, i+5]].copy()
    ta_one_stock.columns = columns
    ta_one_stock = ta_one_stock.dropna(axis = 0)

    
    ta_one_stock['target'] = ta_one_stock['last']>ta_one_stock['open']
    ta_one_stock['bips'] = ta_one_stock['last']/ta_one_stock['open'] - 1
    ta_one_stock['c2o'] = ta_one_stock['open']/(ta_one_stock['close_1d'] + 0.0001)
    
    ta_one_stock['ratio_1d_ma_5d'] = ta_one_stock.close_1d/ta_one_stock.close_1d.rolling(5).mean()
    ta_one_stock['ratio_ma_5d_ma_20d'] = ta_one_stock.close_1d.rolling(5).mean()/ta_one_stock.close_1d.rolling(20).mean()

    ta_one_stock['ratio_volat_5d_volat_20d'] = ta_one_stock.close_1d.rolling(5).std()/ta_one_stock.close_1d.rolling(20).std()
    ta_one_stock['rsi'] = rsi(ta_one_stock['close_1d'])
    
    ta_one_stock = ta_one_stock.dropna(axis = 0)

    data.append(ta_one_stock[columns_used])
```

```python
df = pd.DataFrame(np.concatenate(data), columns = columns_used)
```

# Get train test set

```python
from lightgbm import LGBMClassifier
from sklearn.preprocessing import LabelEncoder

import sys
import heapq
from operator import itemgetter
```

```python
train_features = ['c2o', 'ratio_volat_5d_volat_20d', 'ratio_ma_5d_ma_20d', 'ratio_1d_ma_5d', 'rsi', 'day', 'dayofweek']
df['target'] = LabelEncoder().fit_transform(df['target'])
N = 10
```

```python
dates_train_test = ['2012-01-01', '2013-01-01', '2014-01-01', '2015-01-01', '2016-01-01', 
         '2017-01-01', '2018-01-01', '2019-01-01', '2020-01-01']
```

```python

all_perf = []
dates_backtest = []

for dates_train, dates_test in zip(dates_train_test, dates_train_test[1:]):
    df_train = df[df.date<dates_train]
    df_test = df[(df.date>dates_train)&(df.date<dates_test)].copy()


    model = LGBMClassifier(n_estimators = 20)
    model.fit(df_train[train_features].values, 
              df_train['target'].values)
    pred = model.predict_proba(df_test[train_features].values)[:, 1]
    df_test['probas_up'] = pred


    dates = df_test['date'].sort_values().unique()
    for date in dates:
        subdf = dict(df_test[df_test['date']==date][['probas_up', 'bips']].values)

        topitems = heapq.nlargest(iterable=subdf.items(), key=itemgetter(0), n=N)
        topitemsasdict = dict(topitems)
        perf_long = np.mean(list(topitemsasdict.values()))


        topitems = heapq.nsmallest(iterable=subdf.items(), key=itemgetter(0), n=N)
        topitemsasdict = dict(topitems)
        perf_short = np.mean(list(topitemsasdict.values()))

        all_perf.append((perf_long - perf_short)/2 - 0.0004)
        dates_backtest.append(date)
```

```python
money = pd.DataFrame()
money['dates'] = dates_backtest
money['year'] = money.dates.dt.year
money['all_perf'] = all_perf
```

```python
from keras.layers import Input, BatchNormalization, Dense, Dropout
from keras import regularizers, initializers
from keras import Model
from keras import backend as K
```

```python


def create(nb_numerical_inputs, dropout_rate = 0.4, r_l2 = 0, r_l1 = 0):
    """
    creating architecture of our ANN model
    :return: tensorflow object
    """
    

    numerical_inputs = Input(shape=(nb_numerical_inputs,), name='num')
    numerical_logits = numerical_inputs
    numerical_logits = BatchNormalization()(numerical_logits) #i like the batch normalisation

#deuxieme layer
    numerical_logits = Dense(32,activation='relu', kernel_initializer=initializers.glorot_uniform(seed=23),
                             kernel_regularizer=regularizers.l2(r_l2),
                    activity_regularizer=regularizers.l1(r_l1))(numerical_logits)
    numerical_logits = Dropout(dropout_rate)(numerical_logits)

#troisieme layer
    numerical_logits = Dense(16,activation='relu', kernel_initializer=initializers.glorot_uniform(seed=23),
                             kernel_regularizer=regularizers.l2(r_l2),
                    activity_regularizer=regularizers.l1(r_l1))(numerical_logits)
    numerical_logits = Dropout(dropout_rate)(numerical_logits)

#quatrieme
    numerical_logits = Dense(8,activation='relu', kernel_initializer=initializers.glorot_uniform(seed=23),
                             kernel_regularizer=regularizers.l2(r_l2),
                    activity_regularizer=regularizers.l1(r_l1))(numerical_logits)
    numerical_logits = Dropout(dropout_rate)(numerical_logits)

    logits = Dense(16,activation='relu', kernel_initializer=initializers.glorot_uniform(seed=23),
                   kernel_regularizer=regularizers.l2(r_l2),
                    activity_regularizer=regularizers.l1(r_l1))(numerical_logits)
    logits = Dropout(dropout_rate)(logits)

#Final
    out = Dense(1, activation='sigmoid')(logits)

    model = Model(inputs = numerical_inputs, outputs=out)
    model.compile(optimizer='adam',loss=custom_objective)

    return model
```
